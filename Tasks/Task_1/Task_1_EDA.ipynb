```python
#Country does not carry any significance in the analysis for the dataset
df.drop(columns="Country",inplace=True)
```

```python
#Fixing datatypes for better analysis
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
#Adding some more features for better understanding
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month
df["Order Month Name"] = df["Order Date"].dt.month_name()

df["Shipping Time (Days)"] = (df["Ship Date"] - df["Order Date"]).dt.days

df["Profit Margin"] = df["Profit"] / df["Sales"]
```

```python
# Dropping identifier columns that do not add analytical value
df.drop(
    columns=["Row ID", "Order ID", "Customer ID","Customer Name", "Postal Code"],
    inplace=True
)
```

```python
corr = num_df.corr()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".4f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

```

```python
plt.hist(np.log1p(df["Sales"]), bins=50)
plt.title("Log-Scaled Sales Distribution")
plt.xlabel("log(1 + Sales)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.kdeplot(df["Profit"], fill=True)
plt.axvline(0)
plt.title("Profit Distribution (KDE)")
plt.xlabel("Profit")
plt.show()
```

```python
plt.hist(df["Discount"], bins=20)
plt.title("Distribution of Discount")
plt.xlabel("Discount")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(x="Sales", y="Profit", data=df)
plt.title("Sales vs Profit")
plt.show()
```

```python
sns.boxplot(x="Category", y="Profit", data=df)
plt.title("Profit Distribution by Category")
plt.show()
```

```python
subcat_profit = df.groupby("Sub-Category")["Profit"].mean().sort_values()

subcat_profit.plot(kind="barh")
plt.title("Average Profit by Sub-Category")
plt.xlabel("Average Profit")
plt.show()
```

```python
year_region_profit = df.groupby(
    ["Order Year", "Region"]
)["Profit"].sum().unstack()
year_region_profit.plot(kind="bar")
plt.title("Year-wise Profit by Region")
plt.xlabel("Year")
plt.ylabel("Total Profit")
plt.legend(title="Region")
plt.show()

```

```python
# #The plots clearly indicate that the highest profits corresponds to sales associated with Home Office Segment
segment_profit = df.groupby("Segment")["Profit"].mean().sort_values()

plt.figure(figsize=(8, 4))
plt.hlines(
    y=segment_profit.index,
    xmin=0,
    xmax=segment_profit.values
)
plt.plot(segment_profit.values, segment_profit.index, "o")

plt.xlabel("Average Profit")
plt.title("Average Profit by Segment ")
plt.show()
```

```python
#To find optimum value of discounts to avoid compromising profit , 10% - 20% discount works well
plt.scatter(df["Discount"], df["Profit"], alpha=0.5)
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Discount vs Profit")
plt.show()
```

```python
loss_by_category = df[df["Profit"] < 0].groupby("Category")["Profit"].sum()

plt.figure(figsize=(6, 6))
plt.pie(
    abs(loss_by_category),              
    labels=loss_by_category.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Loss Contribution by Category (%)")
plt.show()

```

```python
category_sales = df.groupby("Category")["Sales"].sum()
(category_sales / category_sales.sum() * 100)

(category_sales / category_sales.sum() * 100).plot(kind="bar")
plt.title("Sales Contribution by Category (%)")
plt.show()
```

```python
subcat_sales = df.groupby("Sub-Category")["Sales"].mean().sort_values()

subcat_sales.plot(kind="barh")
plt.title("Average Sales by Sub-Category")
plt.xlabel("Average Profit")
plt.show()
```

```python
sns.violinplot(x="Region", y="Sales", data=df)
plt.title("Sales Distribution by Region")
plt.show()
```

```python
sales_pct = category_sales / category_sales.sum() * 100

plt.figure(figsize=(6, 6))
plt.pie(
    sales_pct,
    labels=sales_pct.index,
    autopct="%1.1f%%",
    startangle=90
)

centre_circle = plt.Circle((0, 0), 0.70, fc="white")
plt.gca().add_artist(centre_circle)

plt.title("Sales Contribution by Category (%)")
plt.show()
```
